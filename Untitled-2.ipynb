```python
import os
import glob
import pandas as pd

def generate_report(folder_path, output_file, top_x=5):
    """
    Generates a text report of unique values and top X frequent values
    for each column in each Excel file within folder_path.
    
    :param folder_path: The directory containing Excel files
    :param output_file: The path to the output text report
    :param top_x: Number of top frequent values to display per column
    """
    # Find all Excel files in the specified folder (including .xlsx, .xls, etc.)
    excel_files = glob.glob(os.path.join(folder_path, "*.xls*"))
    
    if not excel_files:
        print(f"No Excel files found in {folder_path}.")
        return

    report_lines = []

    for excel_file in excel_files:
        file_name = os.path.basename(excel_file)
        report_lines.append(f"=== File: {file_name} ===\n")
        
        # Attempt to read the Excel file
        try:
            df = pd.read_excel(excel_file)
        except Exception as e:
            report_lines.append(f"Could not read {file_name} due to error: {e}\n\n")
            continue
        
        # If the DataFrame is empty, note it and move on
        if df.empty:
            report_lines.append(f"{file_name} is empty or has no rows.\n\n")
            continue
        
        # Analyze each column
        for col in df.columns:
            report_lines.append(f"--- Column: {col} ---")
            
            # Drop NaN values to avoid counting them in unique/frequency
            column_values = df[col].dropna()
            
            # Get unique values
            unique_vals = column_values.unique()
            report_lines.append(f"Total unique values: {len(unique_vals)}")
            
            # Get top X frequent values
            value_counts = column_values.value_counts()
            top_values = value_counts.head(top_x)
            
            report_lines.append("Top {} most frequent values:".format(top_x))
            for val, count in top_values.items():
                report_lines.append(f"   Value: {val} | Count: {count}")
            
            # Insert a blank line for clarity
            report_lines.append("")
        
        # Insert a separator line between files
        report_lines.append("\n")

    # Write the report to the output file
    with open(output_file, "w", encoding="utf-8") as f:
        f.write("\n".join(report_lines))

    print(f"Report generated: {output_file}")


if __name__ == "__main__":
    # Folder containing Excel files
    FOLDER_PATH = r"D:\Documents\E_Plus_2030_py\lookup_pys"

    # Output report file
    OUTPUT_FILE = "excel_analysis_report.txt"

    # Number of top frequent values to display per column
    TOP_X = 5

    generate_report(FOLDER_PATH, OUTPUT_FILE, TOP_X)
```

```python
import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import Point, LineString
import matplotlib.pyplot as plt
import json

# Function to read CSV files
def read_csv_files():
    lines = pd.read_csv('lines_demo.csv')
    feeders = pd.read_csv('feeders.csv')
    buildings = pd.read_csv('buildings_demo.csv')
    assignments = pd.read_csv('building_assignments.csv')
    
    return lines, feeders, buildings, assignments

# Function to create a node dictionary with coordinates
def create_node_dict(lines, feeders, buildings):
    node_dict = {}
    
    # Add feeders to node_dict
    for _, row in feeders.iterrows():
        node_dict[row['feeder_id']] = {
            'type': 'feeder',
            'lat': row['lat'],
            'lon': row['lon']
        }
    
    # Add buildings to node_dict
    for _, row in buildings.iterrows():
        node_dict[row['building_id']] = {
            'type': 'building',
            'lat': row['lat'],
            'lon': row['lon'],
            'building_type': row['building_type'],
            'peak_load_kW': row['peak_load_kW']
        }
    
    # Add a main substation (estimating its position)
    # We'll place it slightly south-west of the average position of all feeders
    feeder_avg_lat = feeders['lat'].mean()
    feeder_avg_lon = feeders['lon'].mean()
    node_dict['MainSubstation'] = {
        'type': 'substation',
        'lat': feeder_avg_lat - 0.01,  # Offset to place it apart
        'lon': feeder_avg_lon - 0.01
    }
    
    # Add LV branches (they don't have explicit coordinates,
    # so we'll estimate them based on the feeder they're connected to)
    for _, line in lines.iterrows():
        if 'LVbranch' in line['to_id'] and line['to_id'] not in node_dict:
            parent_feeder = line['from_id']
            if parent_feeder in node_dict:
                # Generate a position slightly offset from the parent feeder
                offset_lat = np.random.uniform(-0.003, 0.003)
                offset_lon = np.random.uniform(-0.003, 0.003)
                node_dict[line['to_id']] = {
                    'type': 'lv_branch',
                    'lat': node_dict[parent_feeder]['lat'] + offset_lat,
                    'lon': node_dict[parent_feeder]['lon'] + offset_lon
                }
    
    return node_dict

# Function to create GeoJSON features for nodes
def create_node_features(node_dict):
    features = []
    
    for node_id, node_data in node_dict.items():
        point = Point(node_data['lon'], node_data['lat'])
        
        properties = {
            'id': node_id,
            'type': node_data['type']
        }
        
        # Add additional properties if available
        if 'building_type' in node_data:
            properties['building_type'] = node_data['building_type']
        if 'peak_load_kW' in node_data:
            properties['peak_load_kW'] = node_data['peak_load_kW']
        
        feature = {
            'type': 'Feature',
            'geometry': {
                'type': 'Point',
                'coordinates': [node_data['lon'], node_data['lat']]
            },
            'properties': properties
        }
        
        features.append(feature)
    
    return features

# Function to create GeoJSON features for lines
def create_line_features(lines, node_dict):
    features = []
    
    for _, line in lines.iterrows():
        from_id = line['from_id']
        to_id = line['to_id']
        
        # Skip if any of the nodes doesn't exist in our dictionary
        if from_id not in node_dict or to_id not in node_dict:
            print(f"Warning: Could not find coordinates for line {line['line_id']} ({from_id} to {to_id})")
            continue
        
        # Get coordinates
        from_coords = [node_dict[from_id]['lon'], node_dict[from_id]['lat']]
        to_coords = [node_dict[to_id]['lon'], node_dict[to_id]['lat']]
        
        # Create line geometry
        line_geom = LineString([from_coords, to_coords])
        
        # Create feature
        feature = {
            'type': 'Feature',
            'geometry': {
                'type': 'LineString',
                'coordinates': [from_coords, to_coords]
            },
            'properties': {
                'id': line['line_id'],
                'from_id': from_id,
                'to_id': to_id,
                'length_km': line['length_km'],
                'voltage_level': line['voltage_level']
            }
        }
        
        features.append(feature)
    
    return features

# Function to create GeoJSON features for building connections
def create_building_connection_features(assignments, node_dict, lines):
    features = []
    
    for _, assignment in assignments.iterrows():
        building_id = assignment['building_id']
        line_id = assignment['line_id']
        
        # Skip if building doesn't exist in our dictionary
        if building_id not in node_dict:
            print(f"Warning: Could not find coordinates for building {building_id}")
            continue
            
        # Get line information
        line_info = lines[lines['line_id'] == line_id]
        if line_info.empty:
            print(f"Warning: Could not find line {line_id}")
            continue
            
        # Get the to_id (end point) of the line
        to_id = line_info.iloc[0]['to_id']
        
        # Skip if line endpoint doesn't exist in our dictionary
        if to_id not in node_dict:
            print(f"Warning: Could not find coordinates for line endpoint {to_id}")
            continue
            
        # Get building coordinates and line end coordinates
        building_coords = [node_dict[building_id]['lon'], node_dict[building_id]['lat']]
        line_end_coords = [node_dict[to_id]['lon'], node_dict[to_id]['lat']]
        
        # Create feature
        feature = {
            'type': 'Feature',
            'geometry': {
                'type': 'LineString',
                'coordinates': [building_coords, line_end_coords]
            },
            'properties': {
                'id': f"connection_{building_id}_{line_id}",
                'building_id': building_id,
                'line_id': line_id,
                'distance_km': assignment['distance_km'],
                'connection_type': 'building_connection'
            }
        }
        
        features.append(feature)
    
    return features

# Function to create and save GeoJSON
def create_geojson(node_features, line_features, building_connection_features, output_file='power_network.geojson'):
    # Combine all features
    all_features = node_features + line_features + building_connection_features
    
    # Create GeoJSON structure
    geojson = {
        'type': 'FeatureCollection',
        'features': all_features
    }
    
    # Save to file
    with open(output_file, 'w') as f:
        json.dump(geojson, f)
    
    print(f"GeoJSON saved to {output_file}")
    return geojson

# Function to visualize the network using GeoPandas
def visualize_network(geojson):
    # Convert GeoJSON to GeoDataFrame
    features_df = pd.DataFrame(geojson['features'])
    
    # Separate nodes and lines
    nodes = [f for f in geojson['features'] if f['geometry']['type'] == 'Point']
    lines = [f for f in geojson['features'] if f['geometry']['type'] == 'LineString' and f['properties'].get('connection_type') != 'building_connection']
    building_connections = [f for f in geojson['features'] if f['geometry']['type'] == 'LineString' and f['properties'].get('connection_type') == 'building_connection']
    
    nodes_df = gpd.GeoDataFrame.from_features(nodes)
    lines_df = gpd.GeoDataFrame.from_features(lines)
    connections_df = gpd.GeoDataFrame.from_features(building_connections)
    
    # Create plot
    fig, ax = plt.subplots(figsize=(12, 10))
    
    # Plot lines with different colors based on voltage level
    for voltage in lines_df['voltage_level'].unique():
        subset = lines_df[lines_df['voltage_level'] == voltage]
        color = 'red' if voltage == 'MV' else 'blue'
        subset.plot(ax=ax, color=color, linewidth=2, label=f'{voltage} Lines')
    
    # Plot building connections
    connections_df.plot(ax=ax, color='green', linestyle='--', linewidth=1, alpha=0.7, label='Building Connections')
    
    # Plot nodes with different colors based on type
    node_colors = {
        'substation': 'orange',
        'feeder': 'green',
        'lv_branch': 'purple',
        'building': 'black'
    }
    
    node_sizes = {
        'substation': 100,
        'feeder': 60,
        'lv_branch': 40,
        'building': 30
    }
    
    for node_type, color in node_colors.items():
        subset = nodes_df[nodes_df['type'] == node_type]
        if not subset.empty:
            subset.plot(ax=ax, color=color, markersize=node_sizes[node_type], label=node_type)
    
    # Add labels for important nodes
    for idx, row in nodes_df.iterrows():
        if row['type'] in ['substation', 'feeder', 'lv_branch']:
            ax.annotate(row['id'], xy=(row.geometry.x, row.geometry.y), 
                        xytext=(3, 3), textcoords="offset points")
    
    # Customize plot
    ax.set_title('Power Network Visualization with Building Connections to LV Endpoints')
    ax.legend()
    plt.tight_layout()
    plt.savefig('power_network_visualization.png', dpi=300)
    plt.show()

def main():
    # Read CSV files
    lines, feeders, buildings, assignments = read_csv_files()
    
    # Create node dictionary with coordinates
    node_dict = create_node_dict(lines, feeders, buildings)
    
    # Create GeoJSON features
    node_features = create_node_features(node_dict)
    line_features = create_line_features(lines, node_dict)
    building_connection_features = create_building_connection_features(assignments, node_dict, lines)
    
    # Create and save GeoJSON
    geojson = create_geojson(node_features, line_features, building_connection_features)
    
    # Visualize the network
    visualize_network(geojson)

if __name__ == "__main__":
    main()
```

```python
import csv
import random

def weighted_choice(choices):
    """
    Given a list of (item, weight), randomly select an item, 
    respecting the specified probabilities.
    """
    total = sum(weight for item, weight in choices)
    r = random.uniform(0, total)
    upto = 0
    for item, weight in choices:
        if upto + weight >= r:
            return item
        upto += weight
    return choices[-1][0]  # Fallback, shouldn't happen

def generate_building_data(num_buildings=20):
    """
    Generate synthetic building data, returning a list of dicts. 
    Each dict represents one building's row of data.
    """
    # Example distributions: building_function / building_type
    # You can specify a “percentage” or weight for each category.
    building_function_choices = [
        ("residential", 0.60),
        ("non_residential", 0.40),
    ]
    
    # For 'residential' building_types:
    residential_type_choices = [
        ("Two-and-a-half-story House", 0.3),
        ("Corner House", 0.3),
        ("Apartment", 0.4),
    ]
    
    # For 'non_residential' building_types:
    nonres_type_choices = [
        ("Office Function", 0.5),
        ("Meeting Function", 0.3),
        ("Education Function", 0.2),
    ]
    
    # Example distributions for year_range:
    year_range_choices = [
        ("< 1945", 0.1),
        ("1945 - 1964", 0.1),
        ("1965 - 1974", 0.1),
        ("1975 - 1991", 0.2),
        ("1992 - 2005", 0.3),
        ("2006 - 2014", 0.1),
        ("2015 and later", 0.1)
    ]
    
    # Synthetic ranges for infiltration, occupant density, etc.
    # In real code, you'd look these up from CSV or a param file.
    def random_infiltration_rate(build_func):
        # Perhaps residential has lower infiltration than some non-res types
        if build_func == "residential":
            return (round(random.uniform(0.5, 0.8), 2),
                    round(random.uniform(0.8, 1.0), 2))
        else:
            return (round(random.uniform(0.7, 1.0), 2),
                    round(random.uniform(1.0, 1.3), 2))
    
    def random_occupant_density(build_func):
        # occupant_density_m2_per_person
        # e.g., smaller area/person for apartments vs. large offices
        if build_func == "residential":
            min_val = random.choice([10, 12, 15])
            max_val = min_val + random.randint(5, 10)
            return (min_val, max_val)
        else:
            # non-res could have bigger occupant densities in offices
            min_val = random.choice([5, 8, 10])
            max_val = min_val + random.randint(3, 10)
            return (min_val, max_val)

    def random_dhw_usage(build_func):
        # liters_per_person_per_day
        if build_func == "residential":
            return (random.randint(30, 50), random.randint(50, 70))
        else:
            # Non-res might have very different usage or near zero
            return (random.randint(0, 5), random.randint(5, 15))

    def random_heating_setpoint():
        # e.g. (19,20) to (20,21)
        day_min = random.uniform(19.0, 20.0)
        day_max = day_min + 1.0
        return (round(day_min, 1), round(day_max, 1))

    def random_cooling_setpoint():
        # e.g. (24,25) to (25,26)
        day_min = random.uniform(23.5, 24.5)
        day_max = day_min + 1.0
        return (round(day_min, 1), round(day_max, 1))

    def random_lpd(build_func):
        # lighting_power_density
        if build_func == "residential":
            # Typically lower than offices
            return (round(random.uniform(3, 5), 1),
                    round(random.uniform(5, 7), 1))
        else:
            # Non-res can be higher
            return (round(random.uniform(8, 12), 1),
                    round(random.uniform(12, 16), 1))
    
    buildings = []
    for i in range(num_buildings):
        b_id = f"B{i+1:04d}"  # e.g. B0001, B0002, ...
        build_func = weighted_choice(building_function_choices)
        
        if build_func == "residential":
            b_type = weighted_choice(residential_type_choices)
        else:
            b_type = weighted_choice(nonres_type_choices)
        
        year_range = weighted_choice(year_range_choices)
        inf_min, inf_max = random_infiltration_rate(build_func)
        occd_min, occd_max = random_occupant_density(build_func)
        dhw_min, dhw_max = random_dhw_usage(build_func)
        h_min, h_max = random_heating_setpoint()
        c_min, c_max = random_cooling_setpoint()
        lpd_min, lpd_max = random_lpd(build_func)
        
        # Fake geometry
        area_val = random.randint(30, 300)  # range for area
        perimeter_val = random.randint(20, 100)
        height_val = random.uniform(3.0, 12.0)

        # One row of data
        row = {
            "building_id": b_id,
            "ogc_fid": random.randint(1000000, 9999999),
            "pand_id": f"{random.uniform(1e13, 2e13):.2E}",
            "label": random.choice(["A", "B", "C", "D"]),
            "gem_hoogte": random.randint(2, 8),
            "gem_bouwlagen": random.randint(1, 4),
            "b3_dak_type": random.choice(["flat", "pitched", "multiple horizontal"]),
            "b3_opp_dak_plat": random.randint(20, 100),
            "b3_opp_dak_schuin": random.randint(10, 60),
            "lon": round(random.uniform(4.0, 7.0), 5),
            "lat": round(random.uniform(52.0, 53.5), 5),
            "postcode": f"{random.randint(1000,9999)}{random.choice(['AB','CD','EF','GH','XZ'])}",
            "area": area_val,
            "perimeter": perimeter_val,
            "height": round(height_val, 1),
            "bouwjaar": random.randint(1930, 2023),
            "age_range": year_range,
            "average_wwr": round(random.uniform(0.1, 0.4), 2),
            "building_function": build_func,
            "building_type": b_type,
            
            # Infiltration, occupant density, etc.
            "infiltration_rate_min": inf_min,
            "infiltration_rate_max": inf_max,
            "occupant_density_min": occd_min,
            "occupant_density_max": occd_max,
            "dhw_liters_per_person_day_min": dhw_min,
            "dhw_liters_per_person_day_max": dhw_max,
            "heating_day_setpoint_min": h_min,
            "heating_day_setpoint_max": h_max,
            "cooling_day_setpoint_min": c_min,
            "cooling_day_setpoint_max": c_max,
            "lighting_power_density_wm2_min": lpd_min,
            "lighting_power_density_wm2_max": lpd_max,
            
            # Some random fan power, etc.
            "fan_power_w": random.randint(30, 200),
            "hrv_efficiency": 0.75  # fixed for demonstration
        }
        buildings.append(row)
    
    return buildings


def write_buildings_to_csv(buildings, csv_filename="synthetic_buildings.csv"):
    """
    Write the generated building data to a CSV file.
    """
    # We'll just assume the keys of the first record define the columns:
    fieldnames = list(buildings[0].keys())
    
    with open(csv_filename, mode='w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(buildings)


if __name__ == "__main__":
    # Generate 20 buildings (default)
    building_data = generate_building_data(num_buildings=20)
    write_buildings_to_csv(building_data, csv_filename="synthetic_buildings.csv")
    print("Done! Generated synthetic_buildings.csv with 20 records.")
```
