```python
"""
generate_time_series_loads.py

Generates a time-series CSV (wide format) for multiple buildings and multiple 'Energy' categories.

Output format:

 building_id, Energy, 00:00:00, 00:15:00, 00:30:00, ...
 B0001, heating,  0,   0,   11, ...
 B0001, facility, 121, 121, 0,  ...
 B0001, generation, ...
 B0001, storage, ...
 B0001, total_electricity, ...
 B0002, heating, ...
 B0002, facility, ...
 ...
"""

import csv
import random
from datetime import datetime, timedelta

def generate_time_stamps(
    start_time_str="2025-01-01 00:00:00",
    end_time_str="2025-01-01 06:00:00",
    step_minutes=15
):
    """
    Generate a list of timestamps (as strings HH:MM:SS) from start to end
    in increments of 'step_minutes'.
    Default: from midnight to 6:00 AM in 15-min increments => columns for 6 hours.
    """
    fmt = "%Y-%m-%d %H:%M:%S"
    start_dt = datetime.strptime(start_time_str, fmt)
    end_dt = datetime.strptime(end_time_str, fmt)

    time_stamps = []
    current = start_dt
    while current < end_dt:
        # we only keep the HH:MM:SS part for the CSV header
        time_str = current.strftime("%H:%M:%S")
        time_stamps.append(time_str)
        current += timedelta(minutes=step_minutes)

    return time_stamps

def generate_random_profile(num_steps, category):
    """
    Returns a list of length num_steps with random or patterned values
    depending on the category. You can customize each category's typical range.
    """
    values = []
    if category == "heating":
        # e.g., values ~ 0..30 kW
        for _ in range(num_steps):
            val = round(random.uniform(0, 30), 1)
            values.append(val)
    elif category == "facility":
        # e.g., facility loads ~ 50..150 kW
        for _ in range(num_steps):
            val = round(random.uniform(50, 150), 1)
            values.append(val)
    elif category == "generation":
        # e.g., behind-the-meter solar or something ~ 0..50 kW
        for _ in range(num_steps):
            val = round(random.uniform(0, 50), 1)
            values.append(val)
    elif category == "storage":
        # e.g., battery flow ~ -20..20 kW (negative might be charging, etc.)
        # but let's keep it positive for the example
        for _ in range(num_steps):
            val = round(random.uniform(0, 20), 1)
            values.append(val)
    else:
        # e.g. total_electricity = just sum up the others if you prefer
        # but let's do random again or 0 for now
        for _ in range(num_steps):
            val = 0.0
            values.append(val)

    return values

def generate_time_series_loads(
    building_ids=None,
    categories=("heating", "facility", "generation", "storage", "total_electricity"),
    start_time="2025-01-01 00:00:00",
    end_time="2025-01-01 06:00:00",
    step_minutes=15,
    output_csv="time_series_loads.csv"
):
    """
    Creates a CSV with columns:
      [building_id, Energy, <time_1>, <time_2>, ..., <time_N>]

    Example usage:
      generate_time_series_loads(
         building_ids=["B0001","B0002","B0003"],
         categories=["heating","facility","generation","storage","total_electricity"],
         start_time="2025-01-01 00:00:00",
         end_time="2025-01-01 06:00:00",
         step_minutes=15,
         output_csv="time_series_loads.csv"
      )
    """
    # If building_ids not supplied, let's create a quick list:
    if not building_ids:
        building_ids = [f"B{i+1:04d}" for i in range(3)]  # e.g. B0001,B0002,B0003

    # 1) Generate the list of time stamps
    time_stamps = generate_time_stamps(start_time, end_time, step_minutes)
    num_steps = len(time_stamps)

    # 2) Build the header row
    # The first two columns: building_id, Energy
    # Then each time stamp as a separate column
    header = ["building_id", "Energy"] + time_stamps

    # 3) Prepare the rows:
    rows = []

    # For each building, for each category, we have one row of data
    for b_id in building_ids:
        # generate random or sample data for each category
        # store them in a dictionary so we can combine them for total_electricity if you want
        cat_data = {}
        for cat in categories:
            cat_data[cat] = generate_random_profile(num_steps, cat)

        # If you want "total_electricity" to be the sum:
        # total_profile = [
        #   cat_data["heating"][i]
        #   + cat_data["facility"][i]
        #   - cat_data["generation"][i]
        #   + cat_data["storage"][i]
        #   for i in range(num_steps)
        # ]
        # cat_data["total_electricity"] = [ round(x,1) for x in total_profile ]

        # Now build row for each category
        for cat in categories:
            row = [b_id, cat]
            row.extend(cat_data[cat])  # each time step value
            rows.append(row)

    # 4) Write CSV
    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerows(rows)

    print(f"[generate_time_series_loads] Created '{output_csv}' with {len(rows)} rows.")

# ------------- Standalone Usage -------------
if __name__ == "__main__":
    # Example: 3 buildings, each with 5 categories, over 6 hours in 15-min steps -> 24 columns of time
    # -> CSV with 3 buildings Ã— 5 categories = 15 rows. Each row has 26 columns (2 ID columns + 24 times).
    generate_time_series_loads(
        building_ids=["B0001","B0002","B0009","B0010"],  # can be any list
        categories=["heating","facility","generation","storage","total_electricity"],
        start_time="2025-01-01 00:00:00",
        end_time="2025-01-01 06:00:00",   # end is exclusive
        step_minutes=60,
        output_csv="time_series_loads.csv"
    )
```

```python
"""
main.py

Demonstrates a complete workflow:
  1) Generate buildings (if not already existing)
  2) Determine how many feeders (by building count or load), write feeders.csv
  3) Create MV/LV lines from substation to feeder nodes (read from feeders.csv)
  4) Assign buildings to the nearest LV line
  5) Build a final 'model' dict from these CSV/JSON files
  6) Generate ASCII diagram of the final model
  7) Convert the final model to JSON for the power flow solver (input.json)
  8) Visualize the network as a graph
  9) Run the (dummy) power flow solver -> sym_output.json / asym_output.json
"""

import os
import random
import json

# ---- New modules (pre-steps) ----
from generate_buildings import generate_buildings_table
from determine_num_feeders import determine_feeders
from create_mv_lv_lines import create_mv_lv_lines
from assign_buildings import (
    load_buildings, load_lines,
    assign_buildings_to_lines, write_assignments_csv
)

# ---- Old modules (final steps) ----
from build_network_model import build_network_model
from ascii_generator import generate_ascii_diagram
from json_generator import generate_json_data
from graph_visualizer import visualize_network
from power_flow_solver import solve_power_flow


def main():
    # Step 1) Generate or load building data
    buildings_csv = "buildings_demo.csv"
    if not os.path.exists(buildings_csv):
        print("[main] Generating building data (random example)...")
        distribution = [("Residential", 50), ("Commercial", 30), ("Industrial", 20)]
        generate_buildings_table(
            num_buildings=35,
            building_type_distribution=distribution,
            lat_range=(40.10, 40.20),
            lon_range=(-3.60, -3.50),
            demand_range=(5, 60),
            output_format="csv",
            output_path=buildings_csv
        )
    else:
        print(f"[main] Found existing buildings file: '{buildings_csv}'")

    # Step 2) Determine feeders & write feeders.csv
    feeders_csv = "feeders.csv"
    if not os.path.exists(feeders_csv):
        print("[main] Determining feeders (random approach) -> feeders.csv")
        determine_feeders(
            buildings_csv=buildings_csv,
            feeders_csv=feeders_csv,
            buildings_per_feeder=6,
            placement_mode="random_in_bounding_box",
            lat_buffer=0.01,
            lon_buffer=0.01
        )
    else:
        print(f"[main] Found existing feeders file: '{feeders_csv}'")

    # Step 3) Create MV/LV lines from substation + feeders
    lines_csv = "lines_demo.csv"
    if not os.path.exists(lines_csv):
        print("[main] Creating MV/LV lines -> lines_demo.csv")
        substation_id = "MainSubstation"
        substation_lat = 40.150
        substation_lon = -3.550

        # read feeders
        import csv
        feeder_list = []
        with open(feeders_csv, "r", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                feeder_list.append({
                    "feeder_id": row["feeder_id"],
                    "lat": float(row["lat"]),
                    "lon": float(row["lon"])
                })

        create_mv_lv_lines(
            substation_id=substation_id,
            substation_lat=substation_lat,
            substation_lon=substation_lon,
            feeder_nodes=feeder_list,
            lv_branches_per_feeder=4,
            output_format="csv",
            output_path=lines_csv
        )
    else:
        print(f"[main] Found existing lines file: '{lines_csv}'")

    # Step 4) Assign buildings to nearest LV line
    assignments_csv = "building_assignments.csv"
    if not os.path.exists(assignments_csv):
        print("[main] Assigning buildings -> building_assignments.csv")

        # substation
        substation_id = "MainSubstation"
        substation_lat = 40.150
        substation_lon = -3.550
        node_locations = {substation_id: (substation_lat, substation_lon)}

        # feeders
        import csv
        feeder_list = []
        with open(feeders_csv, "r", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                feeder_list.append(row)

        for f in feeder_list:
            fid = f["feeder_id"]
            node_locations[fid] = (float(f["lat"]), float(f["lon"]))

        # lines
        lines_data = load_lines(lines_csv)
        # replicate logic for LV branches
        for ln in lines_data:
            if ln["voltage_level"] == "LV":
                feeder_id = ln["from_id"]
                lvbranch_id = ln["to_id"]
                if lvbranch_id not in node_locations:
                    parts = lvbranch_id.split("_LVbranch_")
                    if len(parts) == 2:
                        try:
                            branch_i = int(parts[1])
                            f_lat, f_lon = node_locations[feeder_id]
                            offset_lat = f_lat + 0.001 * branch_i
                            offset_lon = f_lon - 0.001 * branch_i
                            node_locations[lvbranch_id] = (offset_lat, offset_lon)
                        except:
                            pass

        # load building & line data
        buildings_data = load_buildings(buildings_csv)

        assignments_list = assign_buildings_to_lines(
            buildings_data,
            lines_data,
            node_locations=node_locations,
            only_lv=True
        )
        write_assignments_csv(assignments_list, assignments_csv)
    else:
        print(f"[main] Found existing building assignments file: '{assignments_csv}'")

    # Step 5) Build final model
    print("[main] Building the unified model dictionary...")
    final_model = build_network_model(
        buildings_path=buildings_csv,
        lines_path=lines_csv,
        assignments_path=assignments_csv,
        node_locations=None
    )

    # Step 6) ASCII diagram
    print("\n[main] Generating ASCII diagram...\n")
    ascii_str = generate_ascii_diagram(final_model)
    print(ascii_str)

    # Step 7) Convert model to JSON for solver input
    print("[main] Converting final model to JSON => 'network_model.json'")
    from json_generator import generate_json_data
    json_data = generate_json_data(final_model)
    with open("network_model.json", "w", encoding="utf-8") as f:
        json.dump(json_data, f, indent=2)

    # Step 8) Visualize
    print("\n[main] Visualizing the network graph...")
    visualize_network(final_model, show_labels=True, title="Distribution Network Graph")

    # Step 9) Run dummy PF solver
    print("[main] Running dummy solver => sym_output.json, asym_output.json")
    solve_power_flow(
        "network_model.json",
        params_json_path=None,
        sym_out_path="sym_output.json",
        asym_out_path="asym_output.json"
    )

    print("[main] Workflow complete. Check sym_output.json and asym_output.json.")


if __name__ == "__main__":
    main()
```

```python
"""
time_series_runner.py

Runs a time-series power flow for each hour (or time column) 
in 'time_series_loads.csv'. Each building has multiple "Energy" categories 
(heating, facility, generation, storage...), but we only use 'total_electricity'.

Outputs results (node voltages, etc.) into a single CSV across all time steps.
"""

import csv
import copy
import json

from build_network_model import build_network_model, update_building_loads
from json_generator import generate_json_data
from power_flow_solver import solve_power_flow_in_memory


def load_time_series_data(ts_file):
    """
    Reads 'time_series_loads.csv' with columns like:
      building_id | Energy | 00:00:00 | 01:00:00 | ...
    We only care about the row where 'Energy' == 'total_electricity'.
    
    Returns:
      load_data = {
         "B0001": [val_t0, val_t1, val_t2, ...],
         "B0002": [...],
         ...
      }
      num_steps = number_of_time_columns
    """
    load_data = {}
    time_headers = []

    with open(ts_file, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f, delimiter=',')  # or delimiter=',' if your file uses commas
        # Example fieldnames: ['building_id', 'Energy', '00:00:00', '01:00:00', '02:00:00', ...]
        headers = reader.fieldnames
        # The time columns start from index 2 onward
        time_headers = headers[2:]

        for row in reader:
            b_id = row["building_id"]
            energy_type = row["Energy"]
            if energy_type.strip().lower() == "total_electricity":
                # build array of floats
                values = []
                for t_col in time_headers:
                    val = float(row[t_col])
                    values.append(val)
                load_data[b_id] = values

    num_steps = len(time_headers)
    return load_data, num_steps, time_headers


def run_time_series_pf(
    buildings_file="buildings_demo.csv",
    lines_file="lines_demo.csv",
    assignments_file="building_assignments.csv",
    ts_file="time_series_loads.csv",
    output_csv="time_series_results.csv"
):
    """
    1) Build a single final model (with build_network_model)
    2) Load time-series data for each building from 'ts_file'
    3) For each time step, override building loads, solve power flow, parse results
    4) Save everything to 'time_series_results.csv'
    """

    print("[time_series_runner] Building base model from CSV files...")
    base_model = build_network_model(
        buildings_path=buildings_file,
        lines_path=lines_file,
        assignments_path=assignments_file,
        node_locations=None
    )

    # 2) Parse time-series load data
    load_data, num_steps, time_headers = load_time_series_data(ts_file)
    print(f"[time_series_runner] Found {num_steps} time steps in {ts_file}.")

    # Prepare storage for final results
    # We'll store rows like: {time_idx, building_id, voltage_pu, p_injection, etc.}
    results_rows = []
    # We'll also want a quick map node_id -> building_name or vice versa
    node_id_to_name = {}
    for nd in base_model["nodes"]:
        node_id_to_name[nd["id"]] = nd["name"]

    # 3) Loop over each time step
    for t_idx in range(num_steps):
        # Copy the model for this time step (so we don't overwrite base_model)
        model_t = copy.deepcopy(base_model)

        # Build a dict of { building_name: p_kW_at_this_time }
        # e.g. p_kW = load_data["B0001"][t_idx]
        this_loads = {}
        for b_name, load_array in load_data.items():
            this_loads[b_name] = load_array[t_idx]

        # 4) Update building loads in model
        update_building_loads(model_t, this_loads)

        # 5) Convert to solver input dict
        from json_generator import generate_json_data
        input_dict = generate_json_data(model_t)

        # 6) Solve power flow *in memory*
        from power_flow_solver import solve_power_flow_in_memory
        pf_results = solve_power_flow_in_memory(input_dict)  # returns {"sym":..., "asym":...}

        # We'll parse the "sym" results to get node voltages, p_injection, etc.
        sym_data = pf_results["sym"]["data"]  # { "node": [...], "line": [...], "shunt": [...] }

        for node_res in sym_data["node"]:
            node_id = node_res["id"]
            voltage_pu = node_res.get("u_pu", None)
            p_inj = node_res.get("p", None)  # in watts if we interpret our dummy solver literally

            # We might only care about building nodes, or we can store all
            building_id = node_id_to_name.get(node_id, "")
            # If building_id starts with "B000", it's a building
            # or you can do a more robust check

            row = {
                "time_step": time_headers[t_idx],  # e.g. "00:00:00"
                "node_id": node_id,
                "node_name": building_id,
                "voltage_pu": voltage_pu,
                "p_injection_w": p_inj
            }
            results_rows.append(row)

    # 7) Write results to CSV
    print(f"[time_series_runner] Writing results to {output_csv}")
    out_headers = ["time_step", "node_id", "node_name", "voltage_pu", "p_injection_w"]
    with open(output_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=out_headers)
        writer.writeheader()
        writer.writerows(results_rows)

    print(f"[time_series_runner] Done. {len(results_rows)} rows written to {output_csv}.")


if __name__ == "__main__":
    """
    Example usage:
      python time_series_runner.py
      (Make sure time_series_loads.csv, buildings_demo.csv, lines_demo.csv, 
       building_assignments.csv exist.)
    """
    run_time_series_pf(
        buildings_file="buildings_demo.csv",
        lines_file="lines_demo.csv",
        assignments_file="building_assignments.csv",
        ts_file="time_series_loads.csv",
        output_csv="time_series_results.csv"
    )
```

```python
"""
time_series_runner.py

Demonstrates a time-series loop that logs BOTH building (node) results
(voltage_pu, p_injection_kW, q_injection_kvar, pf) and line results
(i_from(A), i_to(A), line_rating(A), loading_%).

We output them into a single CSV called time_series_results.csv,
with a column 'record_type' = 'building' or 'line'.
"""

import csv
import copy
from build_network_model import build_network_model, update_building_loads
from json_generator import generate_json_data
from power_flow_solver import solve_power_flow_in_memory


def load_time_series_data(ts_file):
    """
    Reads time_series_loads.csv to get only 'total_electricity' rows per building.
    Returns load_data[b_id][t_idx], plus number of time steps and time headers.
    """
    load_data = {}
    with open(ts_file, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f, delimiter=',')  # or ',' if needed
        headers = reader.fieldnames
        time_cols = headers[2:]  # everything after building_id, Energy

        for row in reader:
            if row["Energy"].strip().lower() == "total_electricity":
                b_id = row["building_id"]
                values = []
                for c in time_cols:
                    values.append(float(row[c]))
                load_data[b_id] = values

    return load_data, len(time_cols), time_cols


def compute_pf_from_pq(p_mw, q_mvar):
    """
    Example function to compute power factor from real power (MW) and reactive power (Mvar).
    PF = P / sqrt(P^2 + Q^2).
    Returns sign consistent with P?
    """
    import math
    p = p_mw
    q = q_mvar
    s = math.sqrt(p**2 + q**2)
    if s < 1e-9:
        return 1.0
    pf = abs(p / s)
    return round(pf, 3)


def run_time_series_pf(
    buildings_file="buildings_demo.csv",
    lines_file="lines_demo.csv",
    assignments_file="building_assignments.csv",
    ts_file="time_series_loads.csv",
    results_csv="time_series_results.csv"
):
    print("[time_series_runner] Build base model once.")
    base_model = build_network_model(
        buildings_path=buildings_file,
        lines_path=lines_file,
        assignments_path=assignments_file
    )

    # We'll store line rating from the model (line["i_n"]) to compute loading.
    # Build a dict line_id -> rating(A)
    line_rating_map = {}
    line_id_to_name = {}
    for ln in base_model["lines"]:
        line_rating_map[ln["id"]] = ln["i_n"]  # nominal current rating from the model
        line_id_to_name[ln["id"]] = ln["name"]

    # Build a mapping node_id-> node_name to identify building vs feeder, etc.
    node_id_to_name = {}
    for nd in base_model["nodes"]:
        node_id_to_name[nd["id"]] = nd["name"]

    # 2) parse time-series loads
    load_data, num_steps, time_headers = load_time_series_data(ts_file)
    print(f"[time_series_runner] Found {num_steps} time columns in {ts_file}.")

    results_rows = []
    # We'll create a single CSV with columns:
    out_headers = [
        "time_step",        # e.g. "00:00:00"
        "record_type",      # 'building' or 'line'
        "building_id",      # if building row
        "line_id",          # if line row
        "voltage_pu",
        "p_injection_kW",
        "q_injection_kvar",
        "pf",
        "i_from_a",
        "i_to_a",
        "line_rating_a",
        "loading_percent"
    ]

    # 3) loop over time steps
    for t_idx in range(num_steps):
        # copy model
        model_t = copy.deepcopy(base_model)

        # build load dict for this time
        loads_this_step = {}
        for b_id, arr in load_data.items():
            loads_this_step[b_id] = arr[t_idx]

        # update building loads
        update_building_loads(model_t, loads_this_step)

        # create solver input
        input_dict = generate_json_data(model_t)

        # in-memory solver
        pf_results = solve_power_flow_in_memory(input_dict)  # returns {"sym":..., "asym":...}

        # parse symmetrical results
        sym_data = pf_results["sym"]["data"]
        # 'node': [ {id, u_pu, p, ...}, ... ]
        # 'line': [ {id, i_from}, ... ]

        # 4) Log building/node info
        for node_res in sym_data.get("node", []):
            node_id = node_res["id"]
            node_name = node_id_to_name.get(node_id, "")
            # assume building if node_name starts with "B"
            if node_name.startswith("B"):
                # building row
                voltage_pu = node_res.get("u_pu", None)
                p_w = node_res.get("p", None)   # in watts
                p_kW = round(p_w/1000.0, 3) if p_w else None
                # q_injection not in dummy solver by default, so let's assume 0 or read from advanced solver
                q_injection_kvar = None
                pf_val = None

                # if you had reactive power, you'd do something like:
                # q_var = node_res.get("q", None)
                # if q_var is not None:
                #    q_injection_kvar = round(q_var / 1000.0, 3)
                #    pf_val = compute_pf_from_pq(p_kW*0.001, q_injection_kvar*0.001)

                row = {
                    "time_step": time_headers[t_idx],
                    "record_type": "building",
                    "building_id": node_name,
                    "line_id": "",
                    "voltage_pu": voltage_pu,
                    "p_injection_kW": p_kW,
                    "q_injection_kvar": q_injection_kvar,
                    "pf": pf_val,
                    "i_from_a": "",
                    "i_to_a": "",
                    "line_rating_a": "",
                    "loading_percent": ""
                }
                results_rows.append(row)

        # 5) Log line info
        for ln_res in sym_data.get("line", []):
            line_id = ln_res["id"]  # matches the solver "id"
            i_from = ln_res.get("i_from", None)  # dummy solver has single i_from
            i_to = None  # dummy solver might not store i_to
            # line rating
            rating_a = line_rating_map.get(line_id, 9999)
            loading_pct = None
            if i_from and rating_a > 0:
                loading_pct = round((i_from / rating_a) * 100, 2)

            row = {
                "time_step": time_headers[t_idx],
                "record_type": "line",
                "building_id": "",
                "line_id": line_id_to_name.get(line_id, f"L_{line_id}"),
                "voltage_pu": "",
                "p_injection_kW": "",
                "q_injection_kvar": "",
                "pf": "",
                "i_from_a": i_from,
                "i_to_a": i_to if i_to else "",
                "line_rating_a": rating_a,
                "loading_percent": loading_pct if loading_pct is not None else ""
            }
            results_rows.append(row)

    # 6) write out a single CSV
    with open(results_csv, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=out_headers)
        writer.writeheader()
        writer.writerows(results_rows)

    print(f"[time_series_runner] Done. Wrote {len(results_rows)} rows to {results_csv}.")


if __name__ == "__main__":
    """
    Example usage:
      python time_series_runner.py
      (Ensure you have time_series_loads.csv, buildings_demo.csv, lines_demo.csv,
       building_assignments.csv in the same folder.)
    """
    run_time_series_pf(
        buildings_file="buildings_demo.csv",
        lines_file="lines_demo.csv",
        assignments_file="building_assignments.csv",
        ts_file="time_series_loads.csv",
        results_csv="time_series_results.csv"
    )
```

```python
"""
time_series_runner_wide.py

Generates a single wide-format CSV with columns:

entity_id, record_type, line_id,
voltage_pu, p_injection_kW, q_injection_kvar, pf,
i_from_a, i_to_a, line_rating_a, loading_percent,
<time step 1>, <time step 2>, ...

Where:
 - If record_type != 'line', then 'entity_id' is the node name (B0001, Feeder1, MainSubstation, etc.)
   and the time-step columns store node voltage in p.u.
   building vs station vs feeder is decided by node name:
     e.g. node_name=="MainSubstation" => 'station'
          node_name.startswith("Feeder") => 'feeder'
          node_name.startswith("B") => 'building'
          else => 'other_node'
   'line_id' is empty for these rows.

 - If record_type='line', we store line_id=the line name, entity_id=''
   The time-step columns store the line current in A.
   We also store final i_from_a, line_rating_a, loading_percent, etc.

We also approximate q_injection_kvar and pf for demonstration, since the dummy solver doesn't provide them.
"""

import csv
import copy
import math

from build_network_model import build_network_model, update_building_loads
from json_generator import generate_json_data
from power_flow_solver import solve_power_flow_in_memory

def load_time_series_data(ts_file):
    """
    Reads time_series_loads.csv with columns:
      building_id, Energy, 00:00:00, 00:15:00, 00:30:00, ...
    We only parse rows where Energy=='total_electricity'.
    Returns load_data[b_id] = [val_t0, val_t1, ...],
    plus a list of time_headers.
    """
    load_data = {}
    time_headers = []

    # Adjust delimiter if needed (',' instead of '\t')
    with open(ts_file, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f, delimiter=',')
        headers = reader.fieldnames
        time_headers = headers[2:]  # from the 3rd column onward

        for row in reader:
            if row["Energy"].strip().lower() == "total_electricity":
                b_id = row["building_id"]
                values = []
                for h in time_headers:
                    values.append(float(row[h]))
                load_data[b_id] = values

    return load_data, time_headers


def get_node_record_type(node_name):
    """
    Logic for deciding if a node is building, station, feeder, or other_node.
    """
    if node_name.startswith("B"):
        return "building"
    elif node_name == "MainSubstation":
        return "station"
    elif node_name.startswith("Feeder"):
        return "feeder"
    else:
        return "other_node"


def run_time_series_pf_wide(
    buildings_file="buildings_demo.csv",
    lines_file="lines_demo.csv",
    assignments_file="building_assignments.csv",
    ts_file="time_series_loads.csv",
    output_csv="time_series_wide.csv"
):
    print("[time_series_runner_wide] Building base model from CSVs...")
    base_model = build_network_model(buildings_file, lines_file, assignments_file)

    # Store line ratings, node names, etc.
    line_rating_map = {}
    line_id_map = {}
    for ln in base_model["lines"]:
        line_rating_map[ln["id"]] = ln["i_n"]  # nominal rating in A
        line_id_map[ln["id"]] = ln["name"]

    node_id_to_name = {}
    for nd in base_model["nodes"]:
        node_id_to_name[nd["id"]] = nd["name"]

    # parse time-series loads
    load_data, time_headers = load_time_series_data(ts_file)
    num_steps = len(time_headers)
    print(f"[time_series_runner_wide] Found {num_steps} time columns => {time_headers}")

    # We'll accumulate time-series arrays for nodes & lines
    node_acc = {}   # node_acc[node_id]["time_values"] = [voltage@t0, voltage@t1, ...]
    node_final = {} # store final-step p_inj, q_inj, etc.

    line_acc = {}
    line_final = {}

    # 1) Loop over time steps, solve PF
    for t_idx in range(num_steps):
        model_t = copy.deepcopy(base_model)

        # build load dict for this time
        step_load = {}
        for b_id, arr in load_data.items():
            step_load[b_id] = arr[t_idx]
        update_building_loads(model_t, step_load)

        # solve PF in memory
        input_dict = generate_json_data(model_t)
        pf_results = solve_power_flow_in_memory(input_dict)
        sym_data = pf_results["sym"]["data"]  # {node:[], line:[], shunt:[]}

        # 2) parse node results
        for nd_obj in sym_data["node"]:
            node_id = nd_obj["id"]
            node_name = node_id_to_name.get(node_id, "")
            v_pu = nd_obj.get("u_pu", 1.0)
            p_w = nd_obj.get("p", 0.0)
            p_kW = p_w/1000.0
            # approximate q, pf
            q_w = 0.3 * p_w
            q_kvar = q_w/1000.0
            s_kW = math.sqrt((p_kW**2)+(q_kvar**2)) if abs(p_kW)>1e-9 or abs(q_kvar)>1e-9 else 1e-9
            pf_val = abs(p_kW/s_kW) if s_kW>1e-9 else 1.0

            if node_id not in node_acc:
                node_acc[node_id] = {"time_values":[None]*num_steps}
                node_final[node_id] = {
                    "p_injection_kW":0.0, "q_injection_kvar":0.0, "pf":1.0, "voltage_pu":1.0
                }
            node_acc[node_id]["time_values"][t_idx] = round(v_pu,4)

            # final
            node_final[node_id]["p_injection_kW"] = round(p_kW,3)
            node_final[node_id]["q_injection_kvar"] = round(q_kvar,3)
            node_final[node_id]["pf"] = round(pf_val,3)
            node_final[node_id]["voltage_pu"] = round(v_pu,3)

        # 3) parse line results
        for ln_obj in sym_data["line"]:
            l_id = ln_obj["id"]
            i_from = ln_obj.get("i_from", 0.0)
            if l_id not in line_acc:
                line_acc[l_id] = {"time_values":[None]*num_steps}
                line_final[l_id] = {
                    "i_from_a":0.0, "line_rating_a":line_rating_map.get(l_id,9999), "loading_percent":0.0
                }
            line_acc[l_id]["time_values"][t_idx] = round(i_from,3)

            rating_a = line_rating_map.get(l_id,9999)
            load_pct = 0.0
            if rating_a>0:
                load_pct = (i_from/rating_a)*100.0
            line_final[l_id]["i_from_a"] = round(i_from,3)
            line_final[l_id]["line_rating_a"] = rating_a
            line_final[l_id]["loading_percent"] = round(load_pct,2)

    # 4) Build final CSV
    # columns
    out_cols = [
        "entity_id","record_type","line_id",
        "voltage_pu","p_injection_kW","q_injection_kvar","pf",
        "i_from_a","i_to_a","line_rating_a","loading_percent"
    ] + time_headers

    rows = []

    # Node rows (buildings, feeders, station, etc.)
    for nd_id, acc in node_acc.items():
        finald = node_final[nd_id]
        node_name = node_id_to_name[nd_id]
        r_type = get_node_record_type(node_name)  # 'building','station','feeder','other_node'

        row = {
            "entity_id": node_name,   # e.g. "B0001", "MainSubstation", "Feeder1"
            "record_type": r_type,
            "line_id": "",
            "voltage_pu": finald["voltage_pu"],
            "p_injection_kW": finald["p_injection_kW"],
            "q_injection_kvar": finald["q_injection_kvar"],
            "pf": finald["pf"],
            "i_from_a": "",
            "i_to_a": "",
            "line_rating_a": "",
            "loading_percent": ""
        }
        # fill time columns with node voltage
        for i, t_col in enumerate(time_headers):
            row[t_col] = acc["time_values"][i]
        rows.append(row)

    # Line rows
    for ln_id, acc in line_acc.items():
        finald = line_final[ln_id]
        row = {
            "entity_id": "",
            "record_type": "line",
            "line_id": line_id_map.get(ln_id, f"L_{ln_id}"),
            "voltage_pu": "",
            "p_injection_kW": "",
            "q_injection_kvar": "",
            "pf": "",
            "i_from_a": finald["i_from_a"],
            "i_to_a": "",   # not in dummy solver
            "line_rating_a": finald["line_rating_a"],
            "loading_percent": finald["loading_percent"]
        }
        # fill time columns with line current
        for i, t_col in enumerate(time_headers):
            row[t_col] = acc["time_values"][i]
        rows.append(row)

    with open(output_csv,"w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=out_cols)
        writer.writeheader()
        for rw in rows:
            writer.writerow(rw)

    print(f"[time_series_runner_wide] Completed. Wrote {len(rows)} rows => {output_csv}")


if __name__=="__main__":
    """
    Example usage:
    python time_series_runner_wide.py
    - We expect time_series_loads.csv, buildings_demo.csv, lines_demo.csv,
      building_assignments.csv in the same folder.
    - Output => time_series_wide.csv
    """
    run_time_series_pf_wide(
        "buildings_demo.csv",
        "lines_demo.csv",
        "building_assignments.csv",
        "time_series_loads.csv",
        "time_series_wide.csv"
    )
```
